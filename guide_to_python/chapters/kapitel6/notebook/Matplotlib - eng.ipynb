# Multiple series in the same plot

```python
x = np.arange(0.0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

Multiple series can be plotted simultaneously by adding them to the **plot()** function argument.

```python
plt.plot(x, y1, 'r-', x, y2, 'b-')
plt.show()
```

# Change series properties

```python
plt.plot(x, y1, 'r-', x, y2, 'b-', linewidth=4.0) # Ändra linjetjocklek för alla serier.
plt.show()
```

```python
line1, line2 = plt.plot(x, y1, 'r-', x, y2, 'b-')
line1.set_linewidth(4.0)
line2.set_linewidth(8.0)
plt.show()
```

```python
lines = plt.plot(x, y1, 'r-', x, y2, 'b-') # lines är en lista av Line2D-objekt
plt.setp(lines, linewidth = 8.0) # Sätter egenskaper för alla linjer i lines
plt.show()
```

# Plotting to multiple figure windows

Several figures can be created using the **figure()** function. Figures can be given numbers, so you can indicate to which figure is the active.

```python
plt.figure(1)
plt.plot(x, y1, 'r-')

plt.figure(2)
plt.plot(x, y2, 'b-')

plt.show()
```

# Subplots

Often, plots must be combined into a single plot. This is done with the **subplots()** function. Arguments to the function is given by a special integer consisting of 3 numbers. The first numbers give the rows and columns for the plot. The last number indicates the ID of the subplot.

211 - [2x1] id = 1

212 - [2x1] id = 2

```python
plt.figure(1)
plt.subplot(211) # Subplot med 2 rader, 1 kolumn och första plotten.
plt.plot(x, y1, 'r-')

plt.subplot(212)
plt.plot(x, y2, 'b-') # Subplot med 2 rader, 1 kolumn och andra plotten.

plt.show()
```

```python
plt.figure(1)
plt.subplot(121) # Subplot med 1 rad, 2 kolumner och första plotten.
plt.plot(x, y1, 'r-')

plt.subplot(122)
plt.plot(x, y2, 'b-') # Subplot med 1 rad, 2 kolumner och andra plotten.

plt.show()
```

# Decorating plots - title and column headers

There are several ways of decorating plots in Matplotlib. All functions in Matplotlib work with the active figure. Switching between figures / subplots is done by using the functions **figure()** and **subplot()**.

```python
plt.figure(1)
plt.subplot(211) # Byt till plot 211
plt.xlabel("x")
plt.ylabel("y1")
plt.plot(x, y1, 'r-')

plt.subplot(212) # Byt till plot 212
plt.xlabel("x")
plt.ylabel("y2")
plt.plot(x, y2, 'b-')

plt.show()
```

Text can be added to a chart using the **text()** function. The text is placed with the same coordinates as for the current plot. LaTeX commands can be used to display mathematical functions in the diagram.

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.text(np.pi/2, 0.8, r'$sin(x=\frac{pi}{2})$', horizontalalignment='center')
plt.plot(x, y, 'r-')
plt.show()
```

Adjusting text alignment is done using the argument *horizontalalignment* to **text()** function.

Annotations can be made using the **annotate()** function.

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.annotate('max', xy=(np.pi/2, 1), xytext=(np.pi/2, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05), horizontalalignment='center'
            )
plt.plot(x, y, 'r-')

plt.show()
```

Axis boundaries can be set with the functions **xlim()** and **ylim()**:

```python
plt.figure(1)
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0.0, np.pi)
plt.ylim(0.0, 1.0)
plt.annotate('max', xy=(np.pi/2, 1), xytext=(np.pi/2, 0.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(x, y, 'r-')
plt.show()
```

The axis scales can be changed from linear to logarithmic using the **xscale()** and **yscale()** functions.

```python
x = np.arange(0,2.0, 0.01)
y = np.exp(x)

plt.figure(1)
plt.yscale('log')
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y, 'r-')

plt.show()
```

## Histogram - hist()

```python
np.random.seed(0)

# exempel data

mu = 100  # medelvärde för distributionen
sigma = 15  # standardavvikelse för distributionen

x = mu + sigma * np.random.randn(437)

num_bins = 50

plt.hist(x, num_bins, density=1)

plt.xlabel('Smarts')
plt.ylabel('Probability density')

plt.show()
```

## Bar charts

```python
np.random.seed(0)

# example data
x = np.arange(20)+1.0
y = np.random.rand(20)

plt.bar(x, y)
plt.xlim(0,21)
plt.xticks(range(1,20+1))

plt.show()
```

## Scatter plots

```python
for color in ['red', 'green', 'blue']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    plt.scatter(x, y, c=color, s=scale, label=color,
                alpha=0.3, edgecolors='none')

plt.legend()   # Lägger till en förklaring i diagrammet
plt.grid(True) # Anger att vi vill ha ett rutnät i diagrammet

plt.show()
```

## Contour plots

```python
N = 100

x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)

z = np.sin(0.5*X*Y)

plt.contourf(X, Y, z, cmap=plt.cm.RdBu) # cmap anger färgskala
plt.colorbar() # Lägger till en färgskala i plotten

plt.show()
```

How the pixels should be softened or interpolated can also be controlled:

```python
# from the docs:

# If interpolation is None, default to rc image.interpolation. See also
# the filternorm and filterrad parameters. If interpolation is 'none', then
# no interpolation is performed on the Agg, ps and pdf backends. Other
# backends will fall back to 'nearest'.
#
# http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.imshow

methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

np.random.seed(0)
grid = np.random.rand(4, 4)

fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

fig.subplots_adjust(hspace=0.3, wspace=0.05)

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(grid, interpolation=interp_method, cmap='viridis')
    ax.set_title(interp_method)

plt.show()
```

## Stream line plots

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U*U + V*V)

fig0, ax0 = plt.subplots()
strm = ax0.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.autumn)
fig0.colorbar(strm.lines)

fig1, (ax1, ax2) = plt.subplots(ncols=2)
ax1.streamplot(X, Y, U, V, density=[0.5, 1])

lw = 5*speed / speed.max()
ax2.streamplot(X, Y, U, V, density=0.6, color='k', linewidth=lw)

fig2, (ax3, ax4) = plt.subplots(ncols=2)
ax3.contourf(X, Y, U, cmap=plt.cm.RdBu)
ax4.contourf(X, Y, V, cmap=plt.cm.RdBu)

plt.show()
```

## 3D plot

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

N = 100

x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)

z = np.sin(0.5*X*Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, z, cmap=plt.cm.RdBu)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_zlim(-5, 5)
plt.show()
```
